Pearson correlation analysis
===

```python
import pandas as pd
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Loading Mars crater dataset

```python
df = pd.read_csv('../datasets/marscrater_pds.csv', low_memory=False)
```

```python
print (scipy.stats.pearsonr(df['DIAM_CIRCLE_IMAGE'], df['DEPTH_RIMFLOOR_TOPOG']))
```

Pearson r = 0.59 reveals statistically significant (p-value = 0.0 < 0.05) above average positive correlation between crater diameter and its depth (the more diameter is, the lower crater bottom resides). This is quite intuitive due to impact nature of the crater formation (the larger asteroid hits the Mars surface, the more diameter it creates with more depth)

```python
g=sns.jointplot(x="DIAM_CIRCLE_IMAGE", y="DEPTH_RIMFLOOR_TOPOG", data=df, kind="reg", xlim=(0,400), ylim=(0,5))
```
